```python
a = np.arange(21) - 10
x = np.arange(-10, 10, 0.1)
y1 = x + 3
y3 = []
for i in a:
    y2 = x*x + 2*x + i + 1
    y3.append(y2)
```

```python
plt.rcParams["figure.figsize"] = 10, 20
plt.plot(x, y1)
for i in xrange(21):
    plt.plot(x, y3[i])
plt.show()
```
